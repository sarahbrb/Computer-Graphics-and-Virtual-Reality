**Courbes de Bézier et polynomes de Bernstein**

```python
import numpy as np
import matplotlib.pyplot as plt

def bezier (P, i ):
    X = i*[0]
    Y = i*[0]
    t = np.linspace(0,1,i)   #Définition des paramètres t pour lesquels on calcul M(t)
    for k in range(i):
        T = t[k]
        X[k] = (1-T)**3*P[0,0] + 3*T*(1-T)**2*P[1,0]+3*T**2*(1-T)*P[2,0]+T**3*P[3,0]   #Calcul de X(t)
        Y[k] = (1-T)**3*P[0,1] + 3*T*(1-T)**2*P[1,1]+3*T**2*(1-T)*P[2,1]+T**3*P[3,1]   #Calcul de Y(t)
    return X,Y
```

```python
def trace_bezier(P,i):
    X,Y = bezier(P,i)
    
    plt.plot(X,Y)
    plt.plot(P[:,0],P[:,1],'o',color='g')
    plt.plot(P[:,0],P[:,1],color='g')
    
    plt.show()
```

```python
P_a = np.array([[0,0], [2,1.5], [3,2], [4,0]])
P_b = np.array([[0,0], [2,1.5], [4,0], [3,2]])
trace_bezier(P_b,250)
trace_bezier(P_a,250)
```
